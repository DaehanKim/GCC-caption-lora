```python
import pandas as pd
import os

df = pd.read_csv("dataset/Train_GCC-training.tsv", sep='\t', header=None, names=['caption', 'image_url'])
df = df.sample(frac=1, random_state=42).reset_index(drop=True)  # 셔플
df = df.iloc[:66643]  
f = open("dataset/train_metadata2.jsonl", 'w')
with jsonlines.Writer(f) as writer:
    for index, row in df.iterrows():
        filename = f"{index+1:06d}.png"
        if os.path.exists(f"dataset/train/{filename}"):
            result = {"file_name":filename,"caption": row['caption']}
            writer.write(result)
f.close()
```

```python
import os
import json

# metadata.jsonl 파일 경로
metadata_path = 'dataset/train/metadata.jsonl'
# 이미지가 저장된 디렉터리
image_dir = 'dataset/train'

# metadata.jsonl 파일에서 파일 이름을 읽어서 집합으로 저장
valid_filenames = set()
with open(metadata_path, 'r') as file:
    for line in file:
        data = json.loads(line)
        valid_filenames.add(data['file_name'])

# 이미지 디렉터리에서 모든 파일을 나열
for filename in os.listdir(image_dir):
    # 파일이 .png 확장자를 가지고 있지만, valid_filenames 집합에 없는 경우
    if filename.endswith('.png') and filename not in valid_filenames:
        file_path = os.path.join(image_dir, filename)
        print(f"Deleting: {file_path}")  # 삭제될 파일명 출력 (선택적)
        os.remove(file_path)  # 파일 삭제

print("Cleanup complete.")
```
